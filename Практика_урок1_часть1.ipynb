```python
# Задать путь к папке проекта и имена файлов
DATASET_PATH = 'C:/Users/Nina/Downloads/Библиотеки Python для Data Science_Юлия Пономарева/Практика к уроку 1/Часть 1/'
FILENAME1 = 'marketing_campaign.csv'
FILENAME2 = 'subscribers.csv'
FILENAME3 = 'users.csv'
```

```python
df1 = pd.read_csv(DATASET_PATH+FILENAME1, sep=',')
display(df1.head(2))
display(f"Количество строк в файле {FILENAME1}: {df1.shape[0]} строка")
```

```python
df2 = pd.read_csv(DATASET_PATH+FILENAME2, sep=',')
display(df2.tail(2))
display(f"Количество строк в файле {FILENAME2}: {df2.shape[0]} строка")
```

```python
df3 = pd.read_csv(DATASET_PATH+FILENAME3, sep=',')
display(df3.sample())
display(f"Количество строк в файле {FILENAME3}: {df3.shape[0]} строка")
```

```python
df1.set_index('user_id', inplace=True)
df2.set_index('user_id', inplace=True)
df3.set_index('user_id', inplace=True)
df_merged = df1.join([df2, df3], how='outer')
df_merged.head()
```

```python
# Можно поробовать оптимизировать объем памяти, выделенный под датасет, изменив тип данных с obect на string
df = df_merged.convert_dtypes()
df.info()
```

Все признаки - категориальные. Значит, в качестве остновных статистик нужно использовать моду и подсчет уникальных значений по каждой категории признака

```python
cols = df_merged.columns
for c in cols:
    print(f"\033[32mКатегории признака {c}")
    display(pd.DataFrame(df[c].value_counts()))
```

### 3. Определить эффективность маркетинговых каналов по привлечению платящих игроков
